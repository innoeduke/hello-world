```python
import pandas as pd
import numpy as np

# import plotly modules
import plotly.express as px
import plotly.graph_objects as go

# cufflinks add plotly chart support to pandas dataframes
import cufflinks as cf
```

### 1. 直接调用 dataframe 来显示图表


```python
# create a dataframe with random numerics using numpy

arr = np.random.randn(50, 4)
df = pd.DataFrame(arr, columns=['A', 'B', 'C', 'D'])

# first show the traditional plot provided by pandas

df.plot()
```

```python
# Do the same with plotly charts. Need to call cufflinks beforehand

cf.go_offline(connected=True)
df.iplot()
```

### 2. 使用 plotly.express 显示图表


```python
# load a dataset from plotly.express built-in dataset

df_stocks = px.data.stocks()
df_stocks.head()
```

```python
# plot a line chart of MSFT stock price

px.line(df_stocks, x='date', y='MSFT', labels={
        'date': 'Date', 'MSFT': 'Price'})
```
